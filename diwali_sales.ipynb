```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt # visualizing data
%matplotlib inline
import seaborn as sns
```

```python
df = pd.read_csv('Diwali Sales Data.csv', encoding= 'unicode_escape')
```

```python
df.shape
```

```python
df.head(5)
```

```python
df.info()
```

```python
df.drop(['Status','unnamed1'],axis=1, inplace=True)
```

```python
df.info()
```

```python
pd.isnull(df).sum()
```

```python
df.dropna(inplace=True)
```

```python
df['Amount'] = df['Amount'].astype('int')
```

```python
df.shape
```

```python
df.columns
```

```python
df.describe()
```

```python
df.rename(columns={'Marital_Status':'Shadi'})
```

```python
Exploratory Data Analysis
```

```python
ax=sns.countplot(x="Gender",data=df)
for bar in ax.containers:
    ax.bar_label(bar)
```

```python
sales_data=df.groupby(['Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.barplot(x="Gender",y="Amount",data=sales_data)
```

```python
 From above graph we see that most buyer are female and amount is also spend more by female.
```

```python
age_group=sns.countplot(data=df,x="Age Group",hue="Gender")
for bar in age_group.containers:
    age_group.bar_label(bar)
```

```python
sale_age=df.groupby(['Age Group'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.barplot(x='Age Group',y='Amount',data=sale_age)
```

```python
From above graph we see that at the age of 26-35 females have spend more amount.
```

```python
ax=sns.countplot(data=df,x='Marital_Status',hue='Gender')
for bar in ax.containers:
    ax.bar_label(bar)
```

```python
sale_data=df.groupby("Marital_Status",as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.barplot(data=sale_data,x='Marital_Status',y='Amount')
```

```python
plt.figure(figsize=(20,5))
ax=sns.countplot(data=df,x='Occupation')
for bar in ax.containers:
    ax.bar_label(bar)
```

```python
plt.figure(figsize=(20,5))
amt=df.groupby(['Occupation'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.barplot(data=amt,x='Occupation',y='Amount')
```

```python
From the above graph we see that most of the buyer are come from the ITsector,Healthcare,Aviation sector and they spend most of the amount.
```

```python
df['Product_Category'].unique()
```

```python
plt.figure(figsize=(20,5))
ax=sns.countplot(data=df,x='Product_Category')
for bar in ax.containers:
    ax.bar_label(bar)
```

```python
p_amt=df.groupby('Product_Category',as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(10)  
plt.figure(figsize=(20,5))
plt.title('Highest selling product')
sns.barplot(data=p_amt,x='Product_Category',y='Amount')
```

```python
From th above graph we see that food,clothing& apparel,electronic& gadgets are highest selling product.                        
```

```python
ax=df.groupby('Product_ID',as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False).head(10)
plt.figure(figsize=(20,5))
plt.title('Most selling Product_id')
sns.barplot(data=ax,x='Product_ID',y='Orders')
```
